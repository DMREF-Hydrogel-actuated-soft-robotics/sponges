```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import seaborn as sns

%matplotlib inline
```

# Fig 2

```python
path='Fig2/'
CSVFiles=glob.glob(path+"*.csv")

plt.figure()
for file in CSVFiles:
    fileName=file.split('\\')[1]
    df=pd.read_csv(file,delimiter=',',header=None)
    plt.plot(df[0],df[1],label=fileName)
plt.legend(bbox_to_anchor=(1.05, 1))
```

# Fig 3

```python
path='Fig3/'
CSVFiles=glob.glob(path+"*.csv")

for file in CSVFiles:
    plt.figure()
    fileName=file.split('\\')[1]
    df=pd.read_csv(file,delimiter=',',header=None)
    plt.plot(df[0],df[1],label=fileName+' - Design A')
    plt.plot(df[0],df[2],label=fileName+' - Design B')
    plt.plot(df[0],df[3],label=fileName+' - Design C')
    plt.plot(df[0],df[4],label=fileName+' - Design D')

    plt.legend(bbox_to_anchor=(1.05, 1))
```

# Fig 4

```python
path='Fig4/'
CSVFiles=glob.glob(path+"*.csv")

# Plotting A
for file in CSVFiles[:1]:
    fileName=file.split('\\')[1]
    df=pd.read_csv(file,delimiter=',',header=None)
    plt.figure()
    plt.plot(df[0],df[1],'o',label=fileName)
    plt.figure()
    plt.plot(df[0],df[2],'o',label=fileName)

plt.figure()

#Plotting B
for file in CSVFiles[1:]:
    fileName=file.split('\\')[1]
    df=pd.read_csv(file,delimiter=',',header=None)
    plt.plot(df[0],df[1],label=fileName)
plt.legend(bbox_to_anchor=(1.05, 1))
```
